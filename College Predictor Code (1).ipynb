```python
import pandas as pd
from pandas import DataFrame
from sklearn import linear_model
import tkinter as tk 
import statsmodels.api as sm
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

f= pd.read_csv(r'college_predictor.csv')
df=DataFrame(f,columns=['GRE Score','TOEFL Score','CGPA','Chance of Admit ']) 

X = df[['GRE Score','CGPA','TOEFL Score']]
Y = df['Chance of Admit '] 
regr = linear_model.LinearRegression()
regr.fit(X, Y)
 
root= tk.Tk() 
 
canvas1 = tk.Canvas(root, width = 1200, height = 450)
canvas1.pack()


symb = tk.PhotoImage(file="cp.gif")
ids = canvas1.create_image(700,200, image=symb)


# New_Interest_Rate label and input box
label1 = tk.Label(root, text='Type GRE Score: ')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)

# New_Unemployment_Rate label and input box
label2 = tk.Label(root, text='Type the CGPA: ')
canvas1.create_window(100, 140, window=label2)

entry2 = tk.Entry (root) # create 2nd entry box
canvas1.create_window(270, 140, window=entry2)

label3 = tk.Label(root, text='Type the TOEFL: ')
canvas1.create_window(100, 180, window=label3)

entry3 = tk.Entry (root) # create 3rd entry box
canvas1.create_window(270, 180, window=entry3)


def values(): 
    global New_GRE_Score #our 1st input variable
    New_GRE_Score= float(entry1.get()) 
    
    global New_CGPA_Score #our 2nd input variable
    New_CGPA_Score = float(entry2.get()) 
    
    global New_TOEFL_Score #our 3rd input variable
    New_TOEFL_Score = float(entry3.get()) 
    
    #Prediction_result  = ('Predicted Chance of Admit: ', str(regr.predict([[New_GRE_Score,New_CGPA_Score,New_TOEFL_Score]])))
    Prediction_result  = ('Predicted Chance of Admit: ', str(regr.predict([[New_GRE_Score,New_CGPA_Score,New_TOEFL_Score]])))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 280, window=label_Prediction)
    
button1 = tk.Button (root, text='Predict Chance of Admit',command=values, bg='orange') # button to call the 'values' command above 
canvas1.create_window(270, 250, window=button1)


#plot 1st scatter 
figure3 = plt.Figure(figsize=(5,4), dpi=100)
ax3 = figure3.add_subplot(111)
ax3.scatter(df['GRE Score'].astype(float),df['Chance of Admit '].astype(float), color = 'r')
scatter3 = FigureCanvasTkAgg(figure3, root) 
scatter3.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax3.legend() 
ax3.set_xlabel('GRE Score')
ax3.set_title('Chance of Admit Vs. GRE Score')

#plot 2nd scatter 
figure4 = plt.Figure(figsize=(5,4), dpi=100)
ax4 = figure4.add_subplot(111)
ax4.scatter(df['CGPA'].astype(float),df['Chance of Admit '].astype(float), color = 'g')
scatter4 = FigureCanvasTkAgg(figure4, root) 
scatter4.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax4.legend() 
ax4.set_xlabel('CGPA')
ax4.set_title('Chance of Admit Vs. CGPA')

#plot 3rd scatter 
figure4 = plt.Figure(figsize=(5,4), dpi=100)
ax4 = figure4.add_subplot(111)
ax4.scatter(df['TOEFL Score'].astype(float),df['Chance of Admit '].astype(float), color = 'b')
scatter4 = FigureCanvasTkAgg(figure4, root) 
scatter4.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
ax4.legend() 
ax4.set_xlabel('TOEFL Score')
ax4.set_title('Chance of Admit Vs. TOEFL Score')



root.mainloop()
```
